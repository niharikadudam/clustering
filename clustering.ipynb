```python
cat_list = []  # List to store names of categorical columns
num_list = []  # List to store names of numerical columns

for i in df.columns:  # Iterate through all columns in the DataFrame
    unique_values = len(df[i].unique())  # Count the number of unique values in the column
    if unique_values < 10:  # If the column has fewer than 10 unique values
        cat_list.append(i)  # Add it to the categorical list
    else:  # Otherwise
        num_list.append(i)  # Add it to the numerical list
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

k = 1
plt.figure(figsize=(12, 12))  # Set figure size
plt.suptitle("Distribution for Numerical Data")  # Add a title for the overall figure

for i in df.loc[:, num_list]:  # Iterate over numerical columns
    plt.subplot(4, 2, k)  # Create a subplot in a 4x2 grid
    sns.distplot(df[i])  # Plot the distribution (updated from sns.distplot)
    plt.title(i)  # Add the column name as the subplot title
    plt.tight_layout()  # Adjust spacing
    k += 1  # Increment subplot index

plt.show()  # Display the plots
```

```python
for i in df.loc[:,num_list]:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3-Q1
    up = Q3 + 1.5*IQR
    low = Q1 - 1.5*IQR

    if df[(df[i] > up) | (df[i] < low)].any(axis=None):
        print(i,"yes")
    else:
        print(i, "no")
```

```python
out_list = ["bonus_trans", "flight_miles_12mo", "flight_trans_12"]

# Remove outliers for the specified columns
for i in out_list:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1
    up_lim = Q3 + 1.5 * IQR
    low_lim = Q1 - 1.5 * IQR
    df.loc[df[i] > up_lim, i] = up_lim  # Cap values greater than upper limit
    df.loc[df[i] < low_lim, i] = low_lim  # Cap values less than lower limit
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

list1 = [2, 3, 4, 5, 6, 7, 8, 9]  # always start number from 2.

for n_clusters in list1:
    clusterer1 = KMeans(n_clusters=n_clusters, random_state=0)
    cluster_labels1 = clusterer1.fit_predict(df)
    sil_score1 = silhouette_score(df, cluster_labels1)
    print(f"Silhouette score for {n_clusters} clusters: {sil_score1}")
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

eps_values = [0.5, 0.7, 0.9, 1.0, 1.2, 1.5]  # Example epsilon values to test
min_samples = 5  # You can adjust this based on your data

for eps in eps_values:
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    cluster_labels = dbscan.fit_predict(df)
    
    # Silhouette score can only be calculated if there are more than 1 cluster
    if len(set(cluster_labels)) > 1:
        sil_score = silhouette_score(df, cluster_labels)
        print(f"Silhouette score for eps={eps}: {sil_score}")
    else:
        print(f"DBSCAN with eps={eps} resulted in a single cluster or noise.")
```
